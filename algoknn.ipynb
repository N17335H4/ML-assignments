```python
import numpy as np
import pandas as pd

#datapoints placed in knn.csv file - read by dataset variable
dataset = pd.read_csv('knn.csv')
dataset
```

```python
#User-input: p
px = 2.5
py = 7
#px = input("Enter x coordinate of input datapoint:")
#py = input("Enter y coordinate of input datapoint:")
```

```python
#Euclidean distances calculated for each datapoint
edist = []
edist = [0 for i in range(17)] 
xarr = np.asfarray(dataset['xpt'])
yarr = np.asfarray(dataset['ypt'])

for i in range(17):
    xd = np.subtract(xarr[i],px)
    xd = round(xd,2)
    xds = np.square(xd)
    yd = np.subtract(yarr[i],py)
    yd = np.round(yd,2)
    yds = np.square(yd)
    sumsq = np.add(xds,yds)
    sq = np.sqrt(sumsq)
    sq = round(sq,2)
    edist[i] = sq

dataset['edistance'] = edist
dataset
```

```python
#find k=3 nearest neighbours
dataset = dataset.sort_values('edistance', ascending = True)
dataset.head(3)
```

```python
#Predict class label
czero = 0
cone = 0
labels = dataset['class'].to_numpy()

for i in range(3):
    if(labels[i]==0): 
        czero = np.add(czero,1)
    else: 
        cone = np.add(cone,1)
    
print("The predicted class for given point (",px,",",py,") is:")
if(czero > cone): 
    print("0")
else: 
    print("1")
```
